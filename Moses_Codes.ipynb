```python
# Function to calculate total, average, and grade of a student
def studentReport(name, mark1, mark2, mark3):
    totalMark = mark1 + mark2 + mark3
    average = totalMark / 3

    # Determine grade
    if average >= 75:
        grade = "A"
    elif average >= 60:
        grade = "B"
    elif average >= 50:
        grade = "C"
    else:
        grade = "F"

    return {
        "name": name,
        "total": totalMark,
        "average": average,
        "grade": grade
    }

# List to store all student reports
students = []

# Collect data for 4 students
for i in range(4):
    print(f"\nEnter details for Student {i + 1}:")
    name = input("Name: ")
    mark1 = int(input("Mark 1: "))
    mark2 = int(input("Mark 2: "))
    mark3 = int(input("Mark 3: "))

    report = studentReport(name, mark1, mark2, mark3)
    students.append(report)

# Display results in a row
print("\n--- Student Reports ---")
for student in students:
    print(f"{student['name']} | Total: {student['total']} | Average: {student['average']:.2f} | Grade: {student['grade']}")

# Find top student
top_student = max(students, key=lambda x: x['average'])

print(f"\nTop Student: {top_student['name']} with an average of {top_student['average']:.2f}")
```

```python
# Handle imbalance : Downsample normal
normal = mydata[mydata['Class'] == 0].sample(n=mydata['Class'].value_counts()[1], random_state=42)
fraud = mydata[mydata['Class'] == 1]
mydata_balanced = pd.concat([normal, fraud]).sample(frac = 1, random_state=42)
```

```python
# Features (X) and Targets (Y)

# All columns except Class
X = mydata_balanced.drop('Class', axis = 1)

# Target: 0 or 1
Y = mydata_balanced['Class']
```

```python
# Split data into train (80 %) and test (20 %)
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Display on the screen
print("Train Shape: ", x_train.shape, "Test Shape:" , x_train.shape)
```

Train the algorithms

```python
# Train the Random Forest Model

model = RandomForestClassifier(n_estimators = 100, random_state=42)

# Fit on train data
model.fit(x_train, y_train)

# Display on the screen
print ("Model Trained.!")
```

```python
# Evaluate the Model

# Predict on test
y_pred = model.predict(x_test)

# Accuracy
accuracy = accuracy_score(y_test, y_pred)

# Display on the screen
print("Accuracy:", accuracy)
```

```python
# Detailed Report : Precision

# Display on the screen
print("\nClassification Report: \n" , classification_report(y_test, y_pred))
```

```python
# 2. LOAD THE DATASET (creditcard.csv from Kaggle)
# ===================================================================
# File has 284,807 transactions, 31 columns: Time, V1-V28 (anonymized features), Amount, Class

df = pd.read_csv('creditcard.csv')        # Make sure file is in your folder!
print("Dataset loaded successfully!")
print(f"Total transactions: {len(df)}")
print(f"Columns: {list(df.columns)}")

# Show first 5 rows
print("\nFirst 5 transactions:")
print(df.head())
```

```python
# 3. EXPLORE THE DATA  VERY IMBALANCED!
# ===================================================================
# Class = 1  Fraud, Class = 0  Normal

print(f"\nClass distribution:")
print(df['Class'].value_counts())

# Calculate percentage
fraud_percent = df['Class'].mean() * 100
print(f"Fraudulent transactions: {fraud_percent:.4f}%  Only ~0.17% are fraud!")

# Visualize class imbalance
plt.figure(figsize=(6, 4))
sns.countplot(x='Class', data=df)
plt.title('Normal (0) vs Fraudulent (1) Transactions', fontsize=10)
plt.xlabel('Class')
plt.ylabel('Count')
plt.yscale('log')  # Log scale to see small fraud count
plt.show()
```

```python
# 4. PREPARE FEATURES AND TARGET
# ===================================================================

# Features = all columns except 'Class'
X = df.drop('Class', axis=1).values       # .values  convert to NumPy array

# Target = fraud label (0 or 1)
y = df['Class'].values

print(f"Feature matrix shape: {X.shape}")  # (284807, 30)
print(f"Target vector shape: {y.shape}")
```

```python
# 5. TRAIN-TEST SPLIT (Keep ratio of fraud/normal same in both)
# ===================================================================

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,              # 20% for testing
    random_state=42,
    stratify=y                  # VERY IMPORTANT: keep same fraud % in train/test
)

print(f"Training samples: {len(X_train)}")
print(f"Testing samples:  {len(X_test)}")
print(f"Fraud in test set: {y_test.sum()} ({y_test.mean()*100:.3f}%)")
```

```python
# 6. FEATURE SCALING  CRITICAL!
# ===================================================================
# 'Amount' and 'Time' are in different scales than V1-V28 (which are already PCA-scaled)

scaler = StandardScaler()
# Fit scaler only on training data (prevent data leakage!)
X_train[:, -2:] = scaler.fit_transform(X_train[:, -2:])  # Scale only Time and Amount
X_test[:, -2:] = scaler.transform(X_test[:, -2:])        # Use same scaler on test

print("Features scaled! Time and Amount now have mean  0 and std  1")
```
